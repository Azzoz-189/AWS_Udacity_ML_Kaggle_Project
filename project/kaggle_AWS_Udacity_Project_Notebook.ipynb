```python
import pandas as pd
from autogluon.tabular import TabularPredictor
```

```python
# Create the train dataset in pandas by reading the csv
# Set the parsing of the datetime column so you can use some of the `dt` features in pandas later
train = pd.read_csv('train.csv', parse_dates=True)
train.head()
```

## Step 3: Train a model using AutoGluon’s Tabular Prediction

Requirements:
* We are predicting `count`, so it is the label we are setting.
* Ignore `casual` and `registered` columns as they are also not present in the test dataset.
* Use the `root_mean_squared_error` as the metric to use for evaluation.
* Set a time limit of 10 minutes (600 seconds).
* Use the preset `best_quality` to focus on creating the best model.

```python
#train.drop(['casual','registered'], inplace=True)
predictor = TabularPredictor(
    label='count',
    eval_metric='root_mean_squared_error',
    learner_kwargs={"ignored_columns": ["casual", "registered"]}).fit(
    train_data=train,
    time_limit=600,
    presets='best_quality')
```

### Review AutoGluon's training run with ranking of models that did the best.

```python
predictor.fit_summary()
```

### Create predictions from test dataset

```python
performance = predictor.evaluate(train)
performance
```

```python
predictions = predictor.predict(test)
predictions.head()
```

#### NOTE: Kaggle will reject the submission if we don't set everything to be > 0.

```python
# Describe the `predictions` series to see if there are any negative values
print(predictions.describe())
```

```python
# How many negative values do we have?
predictions[predictions<0].any()
```

```python
# Set them to zero
predictions = pd.DataFrame([max(0,x) for x in predictions])
predictions[predictions<0].any()
```

### Set predictions to submission dataframe, save, and submit

```python
submission["count"] = predictions
submission.to_csv("submission.csv", index=False)
```

## Step 4: Exploratory Data Analysis and Creating an additional feature
* Any additional feature will do, but a great suggestion would be to separate out the datetime into hour, day, or month parts.

```python
# Create a histogram of all features to show the distribution of each one relative to the data.
#This is part of the exploritory data analysis
train.hist(figsize=(10,10));
```

```python
train['datetime'] = pd.to_datetime(train['datetime'])
test['datetime'] = pd.to_datetime(test['datetime'])
```

```python
# create a new feature
train['month'] = train['datetime'].dt.month
train['day'] = train['datetime'].dt.day
train['hour'] = train['datetime'].dt.hour


test['month'] = test['datetime'].dt.month
test['day'] = test['datetime'].dt.day
test['hour'] = test['datetime'].dt.hour
```

## Make category types for these so models know they are not just numbers
* AutoGluon originally sees these as ints, but in reality they are int representations of a category.
* Setting the dtype to category will classify these as categories in AutoGluon.

```python
train["season"] = pd.Categorical(train["season"])
train["weather"] = pd.Categorical(train["weather"])
test["season"] = pd.Categorical(test["season"])
test["weather"] = pd.Categorical(test["weather"])
```

```python
# View are new feature
train.head()
```

```python
# View histogram of all features again now with the hour feature
train.hist(figsize=(10,10));
```

## Step 5: Rerun the model with the same settings as before, just with more features

```python
predictor_new_features = TabularPredictor(
    label='count',
    eval_metric='root_mean_squared_error',
    learner_kwargs={"ignored_columns": ["casual", "registered"]}).fit(
    train_data=train,
    time_limit=600,
    presets='best_quality')
```

```python
predictor_new_features.fit_summary()
```

```python
performance = predictor_new_features.evaluate(train)
performance
```

```python
# Remember to set all negative values to zero
predictions = predictor_new_features.predict(test)
print(predictions.describe())
predictions = pd.DataFrame([ 0 if predict < 0 else predict for predict in predictions])
print(predictions.describe())
```

## Step 6: Hyper parameter optimization
* There are many options for hyper parameter optimization.
* Options are to change the AutoGluon higher level parameters or the individual model hyperparameters.
* The hyperparameters of the models themselves that are in AutoGluon. Those need the `hyperparameter` and `hyperparameter_tune_kwargs` arguments.

```python
#retrieving the default hyperparameters to know their defualt values:
from autogluon.tabular.configs.hyperparameter_configs import get_hyperparameter_config
from autogluon.tabular.models.tabular_nn.hyperparameters import parameters as nn_hypo
from autogluon.tabular.models.lgb.hyperparameters import parameters as lgb_hypo
from autogluon.tabular.models.catboost.hyperparameters import parameters as cat_hypo
from autogluon.tabular.models.xgboost.hyperparameters import parameters as xgb_hypo
from autogluon.tabular.models.fastainn.hyperparameters import parameters as fastai_hypo
import autogluon.common as ag
```

```python
default_hypers = get_hyperparameter_config('default')
default_hypers
```

```python
#Getting the default model hyperparametrs for each model
nn_hypo_dict = nn_hypo.get_hyper_params(framework='pytorch')
print('NN_Torch ---> ',nn_hypo_dict,'\n')

lgb_hypo_dict = lgb_hypo.get_param_regression_baseline()
print('LGBM ---> ',lgb_hypo_dict,'\n')

cat_hypo_dict = cat_hypo.get_param_regression_baseline()
print('CAT ---> ',cat_hypo_dict,'\n')

xgb_hypo_dict = xgb_hypo.get_param_regression_baseline()
print('XGB ---> ',xgb_hypo_dict,'\n')

fastai_hypo_dcit = fastai_hypo.get_param_regression_baseline()
print('FastAI ---> ',fastai_hypo_dcit,'\n')
```

```python
tuned_hyperparameters={}
```

```python
#starting tuning for tuning parameters:

NN_TORCH =[{'activation': 'relu', 'embedding_size_factor': 1.0, 'embed_exponent': 0.56,
     'max_embedding_dim': 100, 'y_range': None, 'y_range_extend': 0.05, 'dropout_prob': 0.2,
     'optimizer': 'adam', 'learning_rate': 0.003, 'weight_decay': 0.0003,
     'proc.embed_min_categories': 4,'proc.impute_strategy': 'median',
     'proc.max_category_levels': 100,'proc.skew_threshold': 0.99,'use_ngram_features': False,
     'num_layers': 4, 'hidden_size': 256, 'max_batch_size': 64, 'use_batchnorm': False,
     'loss_function': 'auto'}]



tuned_hyperparameters['NN_TORCH']=NN_TORCH
tuned_hyperparameters
```

```python
GBM={
    'num_boost_round':100,
    'num_iterations': 200,
    'num_leaves': ag.space.Int(lower=20, upper=100, default=25),
    'objective':'huber',
    'learning_rate':0.003
    }
tuned_hyperparameters['GBM']=GBM
tuned_hyperparameters
```

```python
XGB =[{'n_estimators': 10000, 'learning_rate': 0.003, 'n_jobs': -1, 'proc.max_category_levels': 100,
   'objective': 'reg:squarederror', 'booster': 'gbtree'}]
tuned_hyperparameters['XGB']=XGB
tuned_hyperparameters
```

```python
CAT=[

    {'iterations': 10000, 'learning_rate': 0.005,'ag_args': {'name_suffix': '_Hypo_1'}}
    ]

tuned_hyperparameters['CAT']=CAT
```

```python

FASTAI=[

     {'layers': None, 'emb_drop': 0.01, 'ps': 0.1, 'bs': 'auto', 'lr': 0.001, 'epochs': 'auto',
      'early.stopping.min_delta': 0.0001, 'early.stopping.patience': 20, 'smoothing': 0.0,
     'ag_args': {'name_suffix': '_Hypo_1'}}
    ]


tuned_hyperparameters['FASTAI']=FASTAI
```

```python

RF=[
    {'learning_rate':0.005, 'ag_args': {'name_suffix': 'lr'}},
    {'criterion': 'squared_error', 'ag_args': {'name_suffix': 'MSE_Hypo_1',
                                                'problem_types': ['regression', 'quantile']}}
    ]

tuned_hyperparameters['RF']=RF
```

```python

XT=[

    {'criterion': 'squared_error',
   'ag_args': {'name_suffix': 'MSE_Hypo_1',
    'problem_types': ['regression', 'quantile']}}
    ]
tuned_hyperparameters['XT']=XT
```

```python
hyperparameters_tune_kwargs= {
    'num_trials': 100, #number of search trials
    'searcher': 'random', #strategy search
    'scheduler': 'local',
    'resource': {'num_cpus':4,'num_gpus':1} #resources
}
```

```python
predictor_new_hpo = TabularPredictor(
    label='count',
    eval_metric='root_mean_squared_error',
    learner_kwargs={"ignored_columns": ["casual", "registered"]}
).fit(train_data=train
      ,time_limit=600,
      presets='best_quality',
      hyperparameters=tuned_hyperparameters
      ,hyperparameter_tune_kwargs=hyperparameters_tune_kwargs)
```
